```python
# OpenCV에서 데이터 열기
# Matplotlib 으로 화면에 출력 
import cv2 
from PIL import Image
import matplotlib.pyplot as plt
# 선 깨짐 방지
%matplotlib inline 

# 이미지 읽어오기
img1 = cv2.imread('./img.jpg')
img2 = Image.open('./img.jpg')
```

```python
# Matplotlib에서 이미지 출력 
plt.imshow(img1)
plt.show()

#-> 이미지 파랗게 보임!
```

```python
plt.imshow(img2)
plt.show()
```

```python
# 이미지 사이즈 확인 
# (가로, 세로, )
# 픽셀 수 = 가로*새로
print(img1.shape) 
print(img2.shape) # 오류! -> 필로우로 읽어들이면 픽셀 획인 불가능ㅠㅠ
```

```python
# 픽셀값 형식으로 이미지 출력 
# 이미지 -> 3차원 구조 / 픽셀수 행단위 ,해상도 열단위 
print(img1) 
print(img2)
```

```python
# 이미지 색상 => RGB 구성 
# 데이터가 불러올떄는 B-G-R 순서로 불러들어와! 

import numpy as np

# 예시 1 (3차원)
user_img = [[[255, 255,255], [0, 0, 0], [255, 255,255]], [[0, 0, 0],[255, 255,255],[0, 0, 0]]]
plt.imshow(user_img)
```

```python
# 예시 2 (2차원)
user_img2 = [[255, 255,255], [0, 0, 0], [255, 255,255]] # 픽셀 1개 
plt.imshow(user_img2)
```

```python
# 예시 3 (3차원)
user_img3 = [[[255, 255,255], [0, 0, 0], [255, 255,255]]] # 한 칸당 픽셀 1개 
plt.imshow(user_img3)
```

```python
# 색 바꾸기 
user_img3[0][0] = [100, 100, 100 ]
print(user_img3)
plt.imshow(user_img3)
```

```python


# 픽셀 단위로 하나의 특정 값만 바꾸기
user_img3[0][0][0] = 255 
print(user_img3)
plt.imshow(user_img3)

user_img3[0][1][1] = 255 
print(user_img3)
plt.imshow(user_img3)

user_img3[0][2][2] = 255 
print(user_img3)
plt.imshow(user_img3)
```

```python
## img1 이미지 컬러로 보기
# BGR -> RGB 순서 바꾸기 
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB ))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB ), cmap='gray')
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY ), cmap='gray')
plt.show()
# BGR2GRAY 로 바꿔주고 뒤에 cmap='gray' 써야 적용 가능 ! 
```

```python
img1_1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB )
print(img1_1[0][0])  # 컬러 이미지 변환 
print(img1[0][0])    # 최초 이미지 

#-> 결과값 바뀌었는데 같아서 안바뀐것 처럼 보임 ! 
```

```python
# 이미지 사이즈 
print(img1.shape)
print(len(img1))       # 행 (세로 사이즈) 
print(len(img1[0]))    # 열 (가로사이즈)
print(len(img1[0][0])) # 컬러 채널 (3가지 색상의 조합)
```

```python
# pandas 모듈을 이용한 배열 이해 
import pandas as pd

b, g, r = cv2.split(img1) # cv2가 제공해주는 기능 
```

```python
cv2.split(img1) 
# 색상들 각각 나눠줌 공백으로 
# ([첫번째 데이터만]), ([두번쨰 데이터만]), ([세번쨰 데이터만]) 모아줌 
```

```python
# 픽셀 배열을 B,G,R 단위로 분할
b, g, r = cv2.split(img1)

# B의 값을 이용해 데이터 프레임 형식으로 변경 
b_df = pd.DataFrame(b)

# 데이터 사이즈 화확인 
print(b_df.shape)

# 데이터 확인 
display(b_df.head())
```

```python
# G의 값을 이용해 데이터 프레임 형식을 변경 
g_df = pd.DataFrame(g)
print(g_df.shape)
display(g_df.head())
```

```python
# R의 값을 이용해 데이터 프레임 형식을 변경 
r_df = pd.DataFrame(r)
print(r_df.shape)
#display(r_df.head())
display(r_df)
```

- 데이터에서 NaN값 제거하기
```
  - df.dropna() : NaN이 1개라도 있는 행 모두 제거
  - df.dropna(axis=1):NaN이 1개라도 있는 열 삭제
  - df.dropna(how="all"): 행 전체가 NaN이면 해당 행 삭제
  - df.dropna(thresh=N개): 각 행에 대하여 N개 이상의 값을 가지는 데이터(행)만 출력 
  - df.dropna(subset=["열이름", "열이름",..]): 지정한 열의 값이 NaN이면 해당 행 삭제
```

- pandas 데이터 병합/연결


```
DataFrame 병합: merge()
```

````python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
````

````python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
````

````python
# 공통 열을 기준으로 병합
pd.merge(df1,df2) 
# 앞(df1)의 "고객번호" 기준으로 서로 매칭
# 
````

````python
pd.merge(df1,df2, how="outer")
# 서로 공통 값 - 연결
# 공통 X 값 -> NaN 
````

````python
# dfq, df2에서 왼쪽에 위차한 데이터(df1)을 기준으로 출력
pd.merge(df1,df2, how="left")
````

````python
# df1, df2에서 오른쪽에 위차한 데이터(df2)을 기준으로 출력
pd.merge(df1,df2, how="right")
````

- DataFrame 데이터 연결 (데이터를 기존 데이터의 아래에 추가할 경우)
  - concat 함수: 여러개 데이터 연결 가능
  - merge -> 2개만 가능


```
df.concat([df1,df2,df3,...]) # 행 방향으로 연결 / 열 제목ㅂ을 기준으로 아래에 추가
df.concat([df1,df2,df3,...], axis=1) # 열 방향으로 연결 / index값을 기준으로 뒤(오른쪽)에 추가
```



````python
df1 = pd.DataFrame({"이름":["홍길동", "이순신", "김파이", "이기준"], 
                    "나이":[30, 45, 29, 48],
                    "성적":[99, 88, 77, 66]})
df1
````

````python
df2 = pd.DataFrame({"이름":["박노트", "이상식", "조규식", "서준석"], 
                    "나이":[24, 88, 68, 59],
                    "성적":[88, 77, 88, 92]})
df2
````

````python
df3 = pd.DataFrame({"이름":["박지정", "김규식", "홍규식", "차준석"], 
                    "나이":[84, 78, 98, 79],
                    "성적":[98, 87, 78, 82],
                    "비고":["간", "폐", "위", "간"]})
df3
````

````python
### 데이터 열 이름 동일한 경우 추가

pd.concat([df1, df2])
# 중복 데이터 존재해도 계속 들어감
# 원래 있던 index 그대로 유지
````

````python
### 필드명 다른 경우 추가

pd.concat([df1, df3])
# "비고" 필드명(열) 추가 
# df1 뒤에 df3 추가 
````

````python
### concat함수 -> 2개 이상의 DataFrame을 한번에 추가 가능 
pd.concat([df1, df2, df3])
````

````python
### concat 열 추가 
df1 = pd.DataFrame([["홍길동", 25,"남"], ["이순신", 48,"남"], ["유관순", 18,"여"]],
                   columns=["이름", "나이", "성별"])
df1              
````

````python
df2 = pd.DataFrame([[85,76,88,"-"],[45,76,65, "관심"],[88,75,91,"-"],[60,60,60,"-"]],
                  columns=["국어", "영어", "수학", "비고"])
df2 = df2.drop([1]) # 1번째 행 삭제
df2
````

````python
pd.concat([df1, df2], axis=1)
# 무조건 index를 기준으로 움직임
# merge -> 일치하는 값을 기준으로 움직임 차이 !! 
# df1 + df2 열 오른쪽으로 추가 / 이름 나이 성별 + 국어 영어 수학 비고

# merge 보다 concat이 더 많이 쓰임! 
# 위에서 1번 데이터 삭제해서 1번 행 NaN값 
````

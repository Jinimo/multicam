```python
from nltk.tokenize import RegexpTokenizer # 정규식 표현에 의한 데이터 전처리 함수
from nltk import Text # text class

import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
```

```python
text='''
n land And now I understand What you tried to say to me How you suffered for yo
you suffered for your sanity How you tried to set them free They would not list
g hand And now I understand What you tried to say to me How you suffered for yo
you suffered for your sanity How you tried to set them free They would not list
snow And now I think I know What you tried to say to me How you suffered for yo
you suffered for your sanity How you tried to set them free They would not list
'''
```

```python
regTok = RegexpTokenizer("[\w]+") # [0-9a-zA-Z_] 영문(대,소)/숫자 제외한 나머지 모두 제거
tokText = Text(regTok.tokenize(text))
print(tokText)
print(type(tokText))
```

```python
# plt.figure(figsize=(14,5)) # 한번에 하나씩 
plt.rcParams['figure.figsize']=(10, 5) # 앞으로의 모든 차트 크기 지정 / 한번만 써주면 다 적용 
plt.title('NLTK Test')
tokText.plot()
plt.show
```

```python
tokText.dispersion_plot(["you", "for", "free", "list"])
```

```python
## 형태소 구분 
import nltk
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
```

```python
nltk.download('book', quiet=True) # 내장된 'book' 영문장 다운로드 
```

```python
from nltk.book import *
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
gute_raw = nltk.corpus.gutenberg.raw('austen-emma.txt')
print(gute_raw[50:985])
```

```python
regTok = RegexpTokenizer("[\w]+") 
tokText = Text(regTok.tokenize(gute_raw[50:985]))

plt.title('NLTK Test')
tokText.plot()
plt.show
```

```python
print(len(gute_raw[50:985]))
tokText.dispersion_plot(["you", "and", "very", "office"])
```

```python
text = gute_raw[50:985]
# ('단어', '품사') 로 보여줌 / 튜플 구조
tag_list = pos_tag(word_tokenize(text))
tag_list
```

```python
print(tag_list[0])    # 첫번째 ('단어','품사')
print(tag_list[0][0]) # '단어'
print(tag_list[0][1]) # '품사'
```

```python
## '품사' = 'NNP' or 'NN' 인 '단어' 데이터만 추출 

nouns_lst = []
for t in tag_list:
    if t[1] == 'NNP' or t[1] == 'NN': # t[1]:단어 
        nouns_lst.append(t[0])
        
print(nouns_lst)
```

```python
## 위 내용 한줄쓰기 표현 (결과 같음)
nouns_lst = [t[0] for t in tag_list if t[1] == 'NNP' or t[1] == 'NN'  ]
print(nouns_lst)
```

```python
# 조건식의 단어 개수
print(len(nouns_lst)) 
```

#### 데이터 갯수 구하기 (pandas)

```python
## 판다스 이용
import pandas as pd

## Series 구조 이용 

# 1개의 얄이 있는 Series 구조 생성 
tokCount = pd.Series(nouns_lst).value_counts()
print(tokCount.head(10)) 
# Series 구조는 DataFrame 구조랑 동일해서 head() 사용 가능!
```

```python
## DataFrame 구조 이용 
tok_df = pd.DataFrame(nouns_lst, columns=['단어'])
tokCount2 = tok_df.groupby('단어')[['단어']].count() # 데이터 갯수 구하기
tokCount2
#tokCount2.columns['빈도수']
#tokCount2.sort_valuse('빈도수', accending=False).head(10)
```

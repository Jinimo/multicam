````python
# 리스트(lst)형 데이터 이용한 DataFrame 생성

import pandas as pd
# 2차원 구조로 만들어져 있어야해!
# 리스트형 -> 한사람의 모든 정보가 1 리스트에 들어감 
lst =[[1, 95, 48, 88],
      [2, 85, 88, 93],
      [3, 80, 70, 60]]

df1 = pd.DataFrame(lst, columns=["번호","국어","영어","수학"]) 
#  columns=[] -> 데이터 열 개수와 동일해야 함
# 행 정의 X
df1
````

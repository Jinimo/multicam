### matplotlip 사용

````python
import matplotlib.pyplot as plt # matpoltlip.pylot -> matpoltlip 안 pylot

a = [1,2,3,4,5,6,] # = y의 위치값들
# 값이 하나만 있을 경우 -> y 위치값 표시

plt.plot(a)
plt.show()
````

````python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5,6]
y = [4,7,6,8,2,9] 

plt.plot(x,y) # x먼저 들어가고 y들어감
plt.show() 
````

````python
x = [1,2,3,4,5,6]
y = [4,7,6,8,2,9] 

plt.plot(x,y, "ro") # "ro" -> 빨간색 점으로 차트 표시 / bo 파란색
plt.show() 
````

````python
y = [4,7,6,8,2,9] 
y2 = [6,9,12,18,22]

# plt.plot() 하나당 하나의 차트 
plt.plot(y)  
plt.plot(y2)

plt.show()
````

````python
y = [4,7,6,8,2,9] 
y2 = [6,9,12,18,22]

# plt.plot() 하나당 하나의 차트 
# plt.plot() 뒤에 plt.show() 붙여주면 해당 차트 보여줌
# 밑은 각 plt.plot()에 대한 plt.show() => 차트 2개
plt.plot(y)  
plt.show() 

plt.plot(y2)
plt.show()
````

````python
### 여러 차트 / 여러 형태 표시 2
x  = [0,1,2,3,4,5]
y1 = [4,7,6,8,2,9] 
y2 = [6,9,12,18,22,14]
y3 = [9,22,3,16,10,15]


plt.plot(x,y1,"r--", x,y2,"bs", x,y3,"g^") 
# (x,y1)를 "r--" 빨간색 점선 (-)
# (x,y2)를 "bs"  파란색 사각 점
# (x,y3)를 "g^"  초록색 삼각 점

plt.show() # (x,y1) (x,y2) (x,y3) 한번에 표시
````

````python
### 여러 차트 / 여러 형태 표시 1
x  = [0,1,2,3,4,5]
y1 = [4,7,6,8,2,9] 
y2 = [6,9,12,18,22,14]
y3 = [9,22,3,16,10,15]


plt.plot(x,y1,"r--", x,y2,"b-.", x,y3,"g:") 
# (x,y1)를 "r--" 빨간색 점선 (-)
# (x,y2)를 "b-."  파란색 점선 (-.)
# (x,y3)를 "g:"  초록색 점선 (..)

plt.show() # (x,y1) (x,y2) (x,y3) 한번에 표시
````

````python
### 여러 차트 / 여러 형태 표시 3
x  = [0,1,2,3,4,5]
y1 = [4,7,6,8,2,9] 
y2 = [6,9,12,18,22,14]
y3 = [9,22,3,16,10,15]


plt.plot(x,y3, "r", label ="y3-data")              # label명 : y3-data
plt.plot(x,y1, "g", marker ="^", label ="y1-data") # marker: 세모 , label명 : y1-data

plt.legend() # 범례 표시 
plt.title( "Graph Title", size=16 ) # 그래프명 , 사이즈 지정

plt.show() # (x,y1) (x,y2) (x,y3) 한번에 표시
````

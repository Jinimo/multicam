time 모듈
- 시간들을 체크해서 보여줌 / 소요된는 시간
- 참조: https://wikidocs.net/33#time
- time.time() : 현재 시간을 실수 형태로 출력 
- time.localtime() : 년, 월, 일, 시, 분, 초 형태로 출력

```python
import time

time.time() # 실수 형태

stime = time.time() # 코드 실행 시작 시간 

print(time.time())
print(time.localtime(time.time())) 
#.strftime 시간->문자 /.strptime 문자->시간
# print(time.strftime("%c", time.localtime(time.time()))) 
print(time.strftime("%x", time.localtime(time.time())))  
# print(time.strftime("%Z", time.localtime(time.time()))) 
# print(time.strftime("%c%Z", time.localtime(time.time()))) 
print("-"*30)

tot = 0
for i in range (100):
  tot += tot + 1

print(tot)

print(time.time()-stime) # 코드 실행 종료 시간 
print("-"*30)

tot = 0
for i in range (100):
  # time.sleep(1) #1초 대기 (** 데이터 처리하는데 시간 걸리니까!, 뒤에서 많이 씀)
  tot += tot + i

print(tot)

print(time.time()-stime)
print("-"*30)

for i in range (10):
  time.sleep(1) #1초 대기 (** 데이터 처리하는데 시간 걸리니까!, 뒤에서 많이 씀)
  tot += tot + i
print(tot)

print(time.time()-stime)
# -> 결과값: 1298074214633706907132624082202613 / 10.012714385986328 -> 10초 기다림
# 나중에 크롤링통해 데이터 모으는데 시간걸릴 경우 사용 !
```

while 문
- https://wikidocs.net/21
- 무한 반복문 
- 몇번 돌려야 끝나는지 모르는 경우에 사용. (끝이 안보이는 경우)
- 잘못하면 무한 루프에 빠질 수 도 있음 
- 조건 True -> 계속 실행
- 조건문을 변경시키거나, if문 이용해서 강제 종료 시켜야 함


```python
# /content/drive/MyDrive/multi/0308 에 폴더 생성
#folder_name = "data_mk"
#os.mkdir(os.getcwd() + "/0308/" + folder_name)
#print(os.listdir(os.getcwd() + "/0308/"))

# 같은 이름의 폴더 중복 생성 X -> 오류뜸
# 기존에 같은 이름의 폴더가 있다면 -> 해당폴더명 + 숫자
folder_name = "data_mk"
fnames = os.listdir(os.getcwd() + "/0308/") # 디렉토리에서 폴더명 가져옴
cnt = 0

while True:
  if folder_name in fnames: # 폴더명 존재 확인 / 새로운 파일 계속 만듬
    cnt +=1                 # cnt(번호) 1씩 증가
    folder_name = folder_name[:7] + str(cnt) # 폴더명 + 증가된 숫자 /  folder_name[:7]->dara_mk 바로 뒤 
  else:
    os.mkdir(os.getcwd() + "/0308/" + folder_name) # 없으면 기존에 있던것에 만들어
    break  # 반복문 강제 종료

print(os.listdir(os.getcwd() + "/0308/"))
```
